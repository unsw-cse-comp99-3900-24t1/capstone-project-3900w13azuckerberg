```python
import pandas as pd
lab_df = pd.read_parquet('raw_data/lab_location.parquet')
lab_df
```

```python
lat_min, lat_max = -43.64, -10.68
long_min, long_max = 113.09, 153.63

# Filter out the rows with latitude and longitude outside the range
filtered_df = lab_df[
    (lab_df['latitude'] < lat_min) | (lab_df['latitude'] > lat_max) |
    (lab_df['longitude'] < long_min) | (lab_df['longitude'] > long_max)
]


# Print the filtered DataFrame
print(filtered_df)
```

```python

def modify_coordinate(lab_df, lab_name, lat, long):
    lab_df.loc[lab_df['lab_name'] == lab_name, 'latitude'] = lat
    lab_df.loc[lab_df['lab_name'] == lab_name, 'longitude'] = long
    return lab_df
    
lab_df = modify_coordinate(lab_df, 'Prince Charles Hospital', -27.39015466564334, 153.0228639106892)
```

```python
# Filter out the rows with latitude and longitude outside the range
filtered_df = lab_df[
    (lab_df['latitude'] < lat_min) | (lab_df['latitude'] > lat_max) |
    (lab_df['longitude'] < long_min) | (lab_df['longitude'] > long_max)
]


# Print the filtered DataFrame
print(filtered_df)
```

```python
inAU = lab_df[
    (lab_df['latitude'] >= lat_min) & (lab_df['latitude'] <= lat_max) &
    (lab_df['longitude'] >= long_min) & (lab_df['longitude'] <= long_max)
]

inAU
```
